```python
from pathlib import Path
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 120)
pd.set_option('display.width', 180)
```

```python
# Locate input files
data_dir = Path('data')
traffic_file = next(data_dir.glob('Traffic_Volume_Counts*(Historical)*.csv'), None)
crash_file = next(data_dir.glob('Motor_Vehicle_Collisions*-Crashes*.csv'), None)

if traffic_file is None or crash_file is None:
    files = [p.name for p in data_dir.glob('*.csv')]
    raise FileNotFoundError(f'Expected both traffic/crash CSV files in {data_dir}. Found: {files}')

print('Traffic file:', traffic_file)
print('Crash file:  ', crash_file)
```

```python
# Load datasets
traffic = pd.read_csv(traffic_file)
crashes = pd.read_csv(crash_file, low_memory=False)

print('traffic shape:', traffic.shape)
print('crashes shape:', crashes.shape)

display(traffic.head(3))
display(crashes.head(3))
```

```python
# ----- Traffic dataset prep -----
traffic = traffic.copy()
traffic.columns = [c.strip() for c in traffic.columns]

# Identify hourly columns based on pattern like '12:00-1:00 AM' or '1:00-2:00PM'
hourly_cols = [
    c for c in traffic.columns
    if re.match(r'^\d{1,2}:\d{2}-\d{1,2}:\d{2}\s?[AP]M$', c.strip())
]

traffic['Date'] = pd.to_datetime(traffic['Date'], errors='coerce')
for c in hourly_cols:
    traffic[c] = pd.to_numeric(traffic[c], errors='coerce')

traffic['daily_volume'] = traffic[hourly_cols].sum(axis=1, min_count=1)
traffic['peak_hour_col'] = traffic[hourly_cols].idxmax(axis=1)

print('Hourly columns detected:', len(hourly_cols))
display(hourly_cols[:8])
display(traffic[['Roadway Name', 'Direction', 'Date', 'daily_volume', 'peak_hour_col']].head())
```

```python
# ----- Crash dataset prep -----
crashes = crashes.copy()
crashes.columns = [c.strip() for c in crashes.columns]

# Convert date/time and key numeric fields
crashes['CRASH DATE'] = pd.to_datetime(crashes['CRASH DATE'], errors='coerce')
crashes['CRASH TIME'] = pd.to_datetime(crashes['CRASH TIME'], format='%H:%M', errors='coerce').dt.time

inj_kill_cols = [
    'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
    'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
    'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
    'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'
]

for c in inj_kill_cols:
    if c in crashes.columns:
        crashes[c] = pd.to_numeric(crashes[c], errors='coerce').fillna(0)

crashes['total_harm'] = crashes.get('NUMBER OF PERSONS INJURED', 0) + crashes.get('NUMBER OF PERSONS KILLED', 0)

display(crashes[['CRASH DATE', 'BOROUGH', 'ON STREET NAME', 'CONTRIBUTING FACTOR VEHICLE 1', 'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'total_harm']].head())
```

```python
# EDA 1: Traffic summary
traffic_summary = pd.DataFrame({
    'rows': [len(traffic)],
    'date_min': [traffic['Date'].min()],
    'date_max': [traffic['Date'].max()],
    'unique_roadways': [traffic['Roadway Name'].nunique() if 'Roadway Name' in traffic.columns else np.nan],
    'avg_daily_volume': [traffic['daily_volume'].mean()]
})
display(traffic_summary)

top_roadways = (traffic.groupby('Roadway Name', dropna=False)['daily_volume']
                .mean()
                .sort_values(ascending=False)
                .head(10))
display(top_roadways.to_frame('avg_daily_volume'))

plt.figure(figsize=(10, 4))
(traffic.groupby('Date')['daily_volume'].sum().sort_index()).plot()
plt.title('Total Daily Traffic Volume (all segments)')
plt.xlabel('Date')
plt.ylabel('Traffic Volume')
plt.tight_layout()
plt.show()
```

```python
# EDA 2: Crash summary
crash_summary = pd.DataFrame({
    'rows': [len(crashes)],
    'date_min': [crashes['CRASH DATE'].min()],
    'date_max': [crashes['CRASH DATE'].max()],
    'boroughs': [crashes['BOROUGH'].nunique(dropna=True) if 'BOROUGH' in crashes.columns else np.nan],
    'total_injuries': [crashes.get('NUMBER OF PERSONS INJURED', pd.Series(dtype=float)).sum()],
    'total_killed': [crashes.get('NUMBER OF PERSONS KILLED', pd.Series(dtype=float)).sum()]
})
display(crash_summary)

if 'BOROUGH' in crashes.columns:
    display(crashes['BOROUGH'].value_counts(dropna=False).head(10).to_frame('crash_count'))

factor_col = 'CONTRIBUTING FACTOR VEHICLE 1'
if factor_col in crashes.columns:
    top_factors = crashes[factor_col].fillna('UNSPECIFIED').value_counts().head(10)
    display(top_factors.to_frame('count'))

plt.figure(figsize=(10, 4))
(crashes.groupby('CRASH DATE').size().sort_index()).plot(color='tomato')
plt.title('Daily Crash Counts')
plt.xlabel('Date')
plt.ylabel('Crashes')
plt.tight_layout()
plt.show()
```

```python
# EDA 3: Compare daily traffic totals vs daily crash counts
traffic_daily = traffic.groupby('Date', as_index=False)['daily_volume'].sum()
crash_daily = crashes.groupby('CRASH DATE', as_index=False).size().rename(columns={'CRASH DATE': 'Date', 'size': 'daily_crashes'})

merged = traffic_daily.merge(crash_daily, on='Date', how='inner').dropna()

print('Overlapping dates:', len(merged))
if len(merged) > 1:
    corr = merged[['daily_volume', 'daily_crashes']].corr().iloc[0, 1]
    print(f'Correlation (daily_volume vs daily_crashes): {corr:.3f}')

display(merged.head())

fig, ax1 = plt.subplots(figsize=(11, 4))
ax1.plot(merged['Date'], merged['daily_volume'], label='Daily Traffic Volume', color='steelblue')
ax1.set_ylabel('Traffic Volume', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')

ax2 = ax1.twinx()
ax2.plot(merged['Date'], merged['daily_crashes'], label='Daily Crashes', color='darkorange', alpha=0.8)
ax2.set_ylabel('Crash Count', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

ax1.set_title('Traffic vs Crashes Over Time (overlapping dates)')
ax1.set_xlabel('Date')
fig.tight_layout()
plt.show()
```
